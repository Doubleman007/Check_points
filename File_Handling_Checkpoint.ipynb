```python
# Importing numpy.
import numpy as np
```

```python
# Assigning the dataset to a variable
filename = "Loan_prediction_dataset.csv"
```

```python
# Using the open() function to open csv file and assigning the result to a variable.
data = open(filename, 'r')
```

```python
# Closing the opened csv file
data.close
```

```python
# Using the delimiter parameter in the genfromtxt() function to specify that the values in the file are separated by commas.
data_set = np.genfromtxt('Loan_prediction_dataset.csv', delimiter = ",", filling_values=0, usecols= 8)
```

```python
data_set
```

```python
# Using the numpy array to perform mean for the loan amount
np.mean(data_set)
```

```python
# # Using the numpy array to perform median for the loan amount
np.median(data_set)
```

```python
# Using the numpy array to perform standard deviation for the loan amount
np.std(data_set)
```
